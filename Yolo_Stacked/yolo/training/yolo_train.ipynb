```python
from ultralytics import YOLO
import os
```

## Preprocess

Since we will not use yolo for classsification merge signature_f and signature_g into one class signature

```python
parent_directory = os.path.join(os.getcwd(), "datasets", "labels")
```

```python
#preprocess all labels 
for child in os.listdir(parent_directory):
    directory = os.path.join(parent_directory, child)
    for name in os.listdir(directory):
        filepath = os.path.join(directory, name)
        with open(filepath, "r") as f:
            lines = f.readlines()

        with open(filepath, "w") as f:
            for line in lines:
                if line.startswith('5') or line.startswith('6'):
                    remaining = line[1:]
                    f.write('0' + remaining)
                    print(f"changed {line} to {'0' + remaining}")
                else:
                    f.write(line)
```

## Train YOLO

```python
#train with default model and export to onnx format
model = YOLO("yolo11n.yaml")
result = model.train(data="ssbi.yaml", epochs=300, batch=-1, patience=10, imgsz=640, device=0)
model.val()
```
